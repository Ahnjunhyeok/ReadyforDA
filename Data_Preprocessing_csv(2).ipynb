* Table join
    - right / left : 겹치는 데이터만 붙이고, 오른쪽 테이블에 있는 정보만 남아있음 / 겹치는 데이터만 붙이고, 왼쪽 테이블에 있는 정보만 남아있음
    - inner : 중복된 데이터만 붙임
    - outer : 중복되지 않은 데이터도 붙임(있는 데이터만 붙임)

```python
df6.head(2)
```

```python
df5
```

```python
df7 = pd.merge(df6,df5, how= "right", on="제품명")
```

```python
df7.to_excel("examples/total_data.xlsx")
```

```python
df7["날짜(전처리)"] = pd.to_datetime(df7["날짜"])
```

```python
df7["공급월"] = df7["날짜(전처리)"].dt.month
```

```python
pd.pivot_table(data=df7, index="공급월", values="재고량", aggfunc="sum") # 피봇팅한 것도 파일 형태로 저장 가능
```

```python
pd.pivot_table(data=df7, index="공급월", values="재고량", aggfunc="sum").plot()
```

```python
sns.barplot(data=df7, x="공급월", y="재고량", ci=None, estimator=sum)
plt.savefig("examples/img_result.png")
```
